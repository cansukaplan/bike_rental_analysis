```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.linear_model import LinearRegression
```

```python
# Load the bike dataset
df = pd.read_csv("uforeports.csv")
df.head(10)
```

```python
# split data target and features
X = df.drop('Time', axis=1)
y = df.Time
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
# Create linear regression object and fit model
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
reg.score(X_test, y_test)
```

```python
predicted = reg.predict(X_test)
predicted
```

```python
result = pd.DataFrame(columns=['target', 'predicted'])
result.target = y_test
result.predicted = predicted
```

```python
result.head(10)
```

```python
mean_absolute_error(result.target, result.predicted)
```
