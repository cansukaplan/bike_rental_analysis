```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
dataset = pd.read_csv("hour.csv",sep=";")
dataset.head(10)
```

```python
X=dataset.drop('mnth',axis=1)
y=dataset.mnth
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
dataset['dteday'] = pd.to_datetime(dataset['dteday']) 
for index, row in dataset.iterrows():

    row['dteday'].tz_localize('UTC').tz_convert('US/Eastern')
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression 
classifier = LogisticRegression(random_state = 0) # random_state = 0 behave as "seed" in random value generator 
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
result = pd.DataFrame(columns=['target', 'predicted'])
result.target = y_test
result.predicted =y_pred
```
