```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression as lr
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("hour.csv",sep=";")
```

```python
x=data["mnth"]
y=data["cnt"]
```

```python
start = 0
for i in range(0, len(y.index)):
    if (i + 1)%10 == 0:
        result2 = y.iloc[start:i+1].values.reshape(10,1)
        start = i + 1
        print (result2)
```

```python
start = 0
for i in range(0, len(x.index)):
    if (i + 1)%10 == 0:
        result1 = x.iloc[start:i+1].values.reshape(10,1)
        start = i + 1
        print (result1)
```

```python
lineerregresyon=lr()
```
